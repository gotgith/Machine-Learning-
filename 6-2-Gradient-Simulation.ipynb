```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_x
```

```python
plot_y = (plot_x-2.5)**2 - 1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
def dJ(theta):
    return 2*(theta - 2.5)
```

```python
def J(theta):
    return (theta - 2.5)**2 - 1
```

```python
eta = 0.1
epsilon = 1e-8

theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break
    
```

```python
theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if(abs(J(theta) - J(last_theta)) < epsilon):
        break

plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color='r', marker='+')
plt.show()
```

```python
def gradient_descent(initial_theta, eta, n_iters=1e4, spsilon=1e-8):
    theta = initial_theta
    theta_history.append(initial_theta)
    i_iters = 0
    
    while i_iters < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_history.append(theta)
    
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
        
        i_iters += 1
```
